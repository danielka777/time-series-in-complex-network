```python
#@title ковертация музыкального файла
!git clone https://github.com/danielka777/cnv2.git &> /dev/null
import sys
sys.path.insert(0,'/content/cnv2')
import функции
from google.colab import files

#sr = 500 #@param {type:"integer"}
частота_дискретизации = 500 #@param {type:"slider", min:0, max:44000, step:500}
sr = частота_дискретизации
alg = "snvg" #@param ["nvg", "hvg", "snvg", "shvg"]

#загрузка файла
uploaded = files.upload()
audio_data = '/content/' + list(uploaded)[0]

#выгрузка аудио
import librosa
x, sr = librosa.load(audio_data, sr=sr)

#слайд
import ipywidgets as widgets
slider = widgets.IntSlider(value=5, min = 1 ,max=len(x))
display(slider)

#кнопка
#from IPython.display import display
button = widgets.Button(description="Click Me!", 
                        button_style='success')

def on_button_clicked(b):
    функции.start(audio_data, sr, alg, slider.value)
    #button.layout.height = '0px'
    button.close()
    slider.close()
    


button.on_click(on_button_clicked)
display(button)


```
